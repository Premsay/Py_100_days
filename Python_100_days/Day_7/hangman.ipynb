```python
import random
word_list = ["aardvark", "baboon", "camel"]

choosen_word = random.choice(word_list)
print(choosen_word)

place = ""

for _ in choosen_word:
    place += "_"
print(place)

guess = input("Guess a letter: ").lower()

display = ""

for letter in choosen_word:
    if letter == guess:
        display += letter
    else:
        display += "_"
print(display)
```

```python
while choosen_word != display:
    guess = input("Guess a letter: ").lower()
    for letter in choosen_word:
        if letter == guess:
            display += letter
        else:
            display += "_"      
    print(display)
```
